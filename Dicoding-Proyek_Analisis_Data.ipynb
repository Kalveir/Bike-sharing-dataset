### Cleaning Data

Dalam tahap pembersihan data, angka-angka pada kolom hari, bulan, musim, hari kerja, dan cuaca akan diubah data kategorikal.

```python

df_day['mnth'] = df_day['mnth'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})

df_day['season'] = df_day['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})

df_day['weekday'] = df_day['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})

df_day['weathersit'] = df_day['weathersit'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})

df_day['yr'] = df_day['yr'].map({
    0: '2011', 1: '2012'
})

df_day['workingday'] = df_day['workingday'].map({
    0: 'Holiday', 1: 'Workingday'
})

# Mengubaha tipe data ke categorical
df_day['season'] = df_day.season.astype('category')
df_day['yr'] = df_day.yr.astype('category')
df_day['mnth'] = df_day.mnth.astype('category')
df_day['holiday'] = df_day.holiday.astype('category')
df_day['weekday'] = df_day.weekday.astype('category')
df_day['workingday'] = df_day.workingday.astype('category')
df_day['weathersit'] = df_day.weathersit.astype('category')

df_day
```

```python

df_hour['mnth'] = df_hour['mnth'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
df_hour['season'] = df_hour['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
df_hour['weekday'] = df_hour['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
df_hour['weathersit'] = df_hour['weathersit'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
df_hour['yr'] = df_hour['yr'].map({
    0: '2011', 1: '2012'
})

df_hour['workingday'] = df_hour['workingday'].map({
    0: 'Holiday', 1: 'Workingday'
})

# Mengubaha tipe data ke categorical
df_hour['season'] = df_hour.season.astype('category')
df_hour['yr'] = df_hour.yr.astype('category')
df_hour['mnth'] = df_hour.mnth.astype('category')
df_hour['holiday'] = df_hour.holiday.astype('category')
df_hour['weekday'] = df_hour.weekday.astype('category')
df_hour['workingday'] = df_hour.workingday.astype('category')
df_hour['weathersit'] = df_hour.weathersit.astype('category')

df_hour
```

## Exploratory Data Analysis (EDA)

## **Tren Penyewaan Sepeda**

```python
trend_bike = df_day.groupby(by='yr').agg({
    'cnt': 'mean'
})

trend_bike.head()
```

Dari hasil di atas, dapat disimpulkan bahwa jumlah penyewaan sepeda meningkat pada tahun 2012 jika dibandingkan dengan tahun 2011.

## **Pengelompokan penyewa sepeda (baik kasual maupun terdaftar) berdasarkan musim**

```python
df_day.groupby('season').agg({
    'casual': 'mean',
    'registered': 'mean',
    'cnt': ['max', 'min', 'mean']
}).reindex (['Spring', 'Summer', 'Fall', 'Winter',])
```

Dari hasil di atas, dapat disimpulkan bahwa jumlah penyewaan sepeda paling tinggi terjadi pada musim gugur, sedangkan paling rendah terjadi pada musim dingin.

## **Pengelompokan penyewa sepeda (baik kasual maupun terdaftar) berdasarkan kondisi cuaca**

Berdasarkan hasil di atas, dapat disimpulkan bahwa jumlah penyewaan sepeda paling tinggi terjadi pada cuaca cerah, sedangkan paling rendah terjadi pada cuaca hujan.

## **Perbandingan penyewa sepeda harian**

```python
df_day.groupby(by='weekday').agg({
    'cnt':['max','min','mean']
}).reindex (['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'])
```

Berdasarkan hasil di atas, dapat disimpulkan bahwa terjadi pola konsisten peningkatan penyewaan sepeda dari hari Senin hingga Jumat. Sebaliknya, terjadi penurunan penyewaan pada hari Sabtu dan Minggu. Hal ini menunjukkan bahwa orang lebih cenderung menyewa sepeda pada hari kerja (Senin hingga Jumat) daripada pada akhir pekan.

## **Perbandingan penyewa sepeda pada hari libur dan hari kerja**

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana tren penyewaan sepeda berubah selama dua tahun (2011-2012)?

```python
trend_bike = trend_bike.reset_index()

trend_bike
```

```python
df_day['mnth'] = pd.Categorical(df_day['mnth'], categories=
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True)

monthly_counts = df_day.groupby(by=["mnth","yr"]).agg({
    "cnt": "sum"
}).reset_index()

sns.lineplot(
    data=monthly_counts,
    x="mnth",
    y="cnt",
    hue="yr",
    palette="rocket",
    marker="o")

plt.title("Trend Sewa Sepeda")
plt.xlabel(None)
plt.ylabel(None)
plt.legend(title="Tahun", loc="upper right")
plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Bagaimana pengaruh kondisi cuaca terhadap jumlah penyewaan sepeda?

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='weathersit',
    y='cnt',
    data=df_day)

plt.title('Jumlah Pengguna Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

### Pertanyaan 3: Bagaimana musim mempengaruhi jumlah pengguna kasual dan terdaftar?

```python
season_pattern = df_day.groupby('season')[['registered', 'casual']].sum().reset_index()

urutan_musim = ['spring', 'summer', 'fall', 'winter']

plt.figure(figsize=(10, 6))

plt.bar(
    season_pattern['season'],
    season_pattern['registered'],
    label='Registered',
    color='tab:red'
)

plt.bar(
    season_pattern['season'],
    season_pattern['casual'],
    label='Casual',
    color='tab:green',
)

plt.xlabel(None)
plt.ylabel(None)
plt.title('Jumlah penyewa sepeda berdasarkan musim')
plt.legend()
plt.show()
```

### Pertanyaan 4: Seberapa besar pengaruh hari libur dan hari kerja terhadap jumlah penyewaan sepeda?

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='workingday',
    y='cnt',
    data=df_day)

plt.title('Perbandingan Penyewa Sepeda Workingday Dan Holiday')
plt.xlabel(None)
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

## Conclusion

## 1. Bagaimana tren penyewaan sepeda berubah selama dua tahun (2011-2012)?
Penyewaan sepeda mengalami peningkatan dari tahun 2011 ke tahun 2012. Selain itu, terjadi peningkatan penyewaan sepeda di pertengahan tahun, dari bulan Mei hingga September, sementara mengalami penurunan pada awal dan akhir tahun.

## 2. Bagaimana pengaruh kondisi cuaca terhadap jumlah penyewaan sepeda?
Pola penyewaan sepeda berdasarkan cuaca menunjukkan bahwa jumlah pengguna sepeda mencapai puncaknya pada cuaca cerah. Di sisi lain, jumlah penyewaan sepeda sedikit menurun pada cuaca berawan, dan mencapai titik terendah pada cuaca hujan.

## 3. Bagaimana musim mempengaruhi jumlah pengguna kasual dan terdaftar?
Pola penyewaan sepeda berdasarkan musim menunjukkan peningkatan bertahap dari musim ke musim. Dimulai dari musim semi, yang memiliki jumlah penyewaan sepeda paling rendah, penyewaan sepeda kemudian meningkat pada musim panas dan gugur yang menjadi musim dengan penyewaan sepeda terbanyak. Jumlah penyewaan sepeda kemudian sedikit menurun pada musim dingin.

## 4. Seberapa besar pengaruh hari libur dan hari kerja terhadap jumlah penyewaan sepeda?
Penyewaan sepeda mencapai puncaknya pada hari kerja, menunjukkan bahwa sepeda digunakan untuk kegiatan sehari-hari seperti pergi bekerja, sekolah, belanja, dan kegiatan lainnya.
